## The Guest Experience at Walt Disney World
Ever since Walt Disney World reopened to the public after the COVID-19 pandemic shut down, there seems to have been an uptick of guest complaints regarding cast members and the parks “losing the magic”. This is something that I deal with every day as a frontline cast member, and I want to ensure that the people who visit the parks have a memorable experience. There are a lot of intricate details that go into the planning of a guest’s day; ride wait times, park hours, amount of people in the park, the time of year that the guest is visiting, etc.  Because there are so many details, many people struggle with finding the appropriate time to take their vacations. For example, a guest may want to experience cooler Florida weather, so they decide to visit in December; however, December is a peak season due to the holidays, so the guest may become upset that they have to wait in longer attraction lines. On the other hand, there are some times of the year where not many people visit at all, and Disney may need to provide an incentive to draw in a crowd. Typically, that incentive is lower prices, which makes Disney more affordable for the average family. Ultimately, my goal with this data analysis is to answer the question: “How can a guest best optimize their vacation in the Walt Disney World Parks?”. 
   The data set I chose provides insight to past trends during different points of the year. It categorizes data by “peak” seasons, the date and year, and even by special events and the weather of that specific day. With the plethora of data provided, I am able to analyze park attendance and different factors that may affect how many people visit a Walt Disney World park.

Graph 1 shows that during peak seasons, Magic Kingdom keeps their gates open as long as they can to maximize the guest experience. Graph 2 shows that during peak seasons, the other parks are not open as long as Magic Kingdom is open. If a guest wants to experience as much Disney as they can in one day, they are better off booking their tickets for a full day at Magic Kingdom IF length of time in the park is the most important thing to them.

Graph 3 shows mean temperatures during different months of the year. If a guest wants to visit during a colder month, they are better off going during months where the average tempature is low, near 40 degrees. Graph 4 shows that throughout 2019, MK changed capacity to meet peak times of the year. On roughly days 50-100, capacity was at its peak. Then, during slower months, capacity was limited and not as many people were let in. This graph would appease a guest whose main concern in the amount of people in the parks.

I am now splitting the data into a test and train set. I am going to do this based on the column WEEKOFYEAR, since many vacations span from a few days to weeks.

```python
from sklearn.model_selection import train_test_split
```

```python
# I am getting an error when running the model, so I am going to go through and remove infinite or NaN values (again)
df_dummies.replace([np.inf, -np.inf], np.nan, inplace=True)
df_dummies.fillna(0, inplace=True)
```

```python
# Splitting into new dataset with focus on column WEEKOFYEAR.
split_data = df_dummies

y = split_data.WEEKOFYEAR
x = split_data.drop('WEEKOFYEAR',axis=1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y)
```

I am going to run a linear regression model first to see if the model is the best fit for my data. I think decision tree may work best, but I want to be able to compare the results between the two.

```python
from sklearn.linear_model import LinearRegression
lrm = LinearRegression()
```

```python
lrm.fit(x_train,y_train)
print(lrm.predict(x_test))
print("Accuracy: ",lrm.score(x_test,y_test))
```

I am also going to include the R2 value and RSME on the test set to see if this model is the best fit.

```python
from sklearn.metrics import r2_score
from sklearn import metrics
```

```python
# Test Set

y_test_pred = lrm.predict(x_test)

# R2
r_squared_test = r2_score(y_test, y_test_pred) 
print('Coefficient of Determination for test set: ', r_squared_test) 

# RMSE
rmse_test = metrics.mean_squared_error(y_test, y_test_pred, squared=False)
print("RMSE of test set: ",rmse_test)
```

According to https://www.researchgate.net/post/Whats-the-acceptable-value-of-Root-Mean-Square-Error-RMSE-Sum-of-Squares-due-to-error-SSE-and-Adjusted-R-square#:~:text=Based%20on%20a%20rule%20of,more%20is%20acceptable%20as%20well:
"Based on a rule of thumb, it can be said that RMSE values between 0.2 and 0.5 shows that the model can relatively predict the data accurately. In addition, Adjusted R-squared more than 0.75 is a very good value for showing the accuracy. In some cases, Adjusted R-squared of 0.4 or more is acceptable as well."

The RSME is close to the 0.2-0.5 range, and the accuracy above was .99. I am going to continue with the decision tree model and see if the numbers I get differ.

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
```

```python
clf.fit(x_train,y_train)
y_pred = clf.predict(x_test)
print("Accuracy:", clf.score(x_test,y_test))
```

```python
# R2
r_squared_test2 = r2_score(y_test, y_pred) 
print('Coefficient of Determination for test set: ', r_squared_test2) 

# RMSE
rmse_test2 = metrics.mean_squared_error(y_test, y_pred, squared=False)
print("RMSE of test set: ",rmse_test2)
```

Between the two models I've chosen, I think that Decision Tree is still the best route to go; despite the accuracy of the linear regression model being 0.99.
